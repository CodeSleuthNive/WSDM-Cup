```python
import pandas as pd
from textblob import TextBlob
from langdetect import detect
import nltk
from nltk.tokenize import word_tokenize
from sklearn.feature_extraction.text import CountVectorizer
from transformers import pipeline
from collections import Counter
import string

# Load the data (assuming the data is in a CSV format)
data = df

# Initialize NLP tools
nltk.download('punkt')

import nltk
nltk.download('punkt_tab')




# Sentiment analysis model
sentiment_analyzer = pipeline("sentiment-analysis")

# Function to compute text sentiment (positive/neutral/negative)
def get_sentiment(text):
    blob = TextBlob(text)
    sentiment = blob.sentiment.polarity
    return sentiment

# Function to detect language (default is using langdetect)
def detect_language(text):
    try:
        return detect(text)
    except:
        return "unknown"

# Function to compute lexical diversity (number of unique words / total words)
def lexical_diversity(text):
    words = word_tokenize(text)
    unique_words = set(words)
    return len(unique_words) / len(words) if len(words) > 0 else 0

# Function to compute sentence complexity (average number of words per sentence)
def sentence_complexity(text):
    sentences = nltk.sent_tokenize(text)
    total_words = sum(len(word_tokenize(sentence)) for sentence in sentences)
    return total_words / len(sentences) if len(sentences) > 0 else 0

# Function to extract features from the dataset
def extract_features(data):
    features = []

    for _, row in data.iterrows():
        id = row['id']
        prompt = row['prompt']
        response_a = row['response_a']
        response_b = row['response_b']

        # Feature extraction for prompt
        prompt_len = len(prompt.split())
        prompt_sentiment = get_sentiment(prompt)
        # prompt_language = detect_language(prompt)

        # Feature extraction for Response A
        response_a_len = len(response_a.split())
        response_a_sentiment = get_sentiment(response_a)
        # response_a_language = detect_language(response_a)
        response_a_lexical_diversity = lexical_diversity(response_a)
        response_a_sentence_complexity = sentence_complexity(response_a)

        # Feature extraction for Response B
        response_b_len = len(response_b.split())
        response_b_sentiment = get_sentiment(response_b)
        # response_b_language = detect_language(response_b)
        response_b_lexical_diversity = lexical_diversity(response_b)
        response_b_sentence_complexity = sentence_complexity(response_b)

        # Store the features for each row
        features.append({
            'id':id,
          'prompt': prompt,
            'prompt_len': prompt_len,
            'prompt_sentiment': prompt_sentiment,
            # 'prompt_language': prompt_language,
            'response_a_len': response_a_len,
            'response_a_sentiment': response_a_sentiment,
            # 'response_a_language': response_a_language,
            'response_a_lexical_diversity': response_a_lexical_diversity,
            'response_a_sentence_complexity': response_a_sentence_complexity,
            'response_b_len': response_b_len,
            'response_b_sentiment': response_b_sentiment,
            # 'response_b_language': response_b_language,
            'response_b_lexical_diversity': response_b_lexical_diversity,
            'response_b_sentence_complexity': response_b_sentence_complexity,
            'winner': row['winner']
        })

    # Convert the features list to a DataFrame
    feature_df = pd.DataFrame(features)
    return feature_df


# Then run the feature extraction as before
feature_df = extract_features(data)

# Save the features to a new CSV (optional)
# feature_df.to_csv('extracted_features.csv', index=False)

# Example output

```
